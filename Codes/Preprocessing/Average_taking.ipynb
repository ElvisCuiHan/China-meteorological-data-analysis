```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("2019-O3_8h.csv")
a=np.argmax(data.columns == "北京")
b=np.argmax(data.columns == "上海")
c=np.argmax(data.columns == "广州")
d=np.argmax(data.columns == "杭州")
e=np.argmax(data.columns == "西安")
data = data.iloc[:,[0,1,2,a,b,c,d, e]]

mean_value = []
for i in range(12):
    hi = np.array(data['date'] < (20190000 + (i + 2) * 100)) & \
            np.array(data['date'] > 20190000 + (i + 1) * 100)
    mean_value.append(np.array(data[hi].iloc[:, 3:].mean().T))

mean_value = pd.DataFrame(mean_value)
mean_value.columns = data.columns[3:]

month = list(range(1, 13))
mean_value.insert(0, "Month", month)
mean_value.insert(1, "type", "O3_8h")
mean_value.head()
```

```python
data = pd.read_csv("2019-PM2.5_raw.csv")
a=np.argmax(data.columns == "北京")
b=np.argmax(data.columns == "上海")
c=np.argmax(data.columns == "广州")
d=np.argmax(data.columns == "杭州")
e=np.argmax(data.columns == "西安")
data = data.iloc[:,[0,1,2,a,b,c,d,e]]

PM = []
for i in range(12):
    hi = np.array(data['date'] < (20190000 + (i + 2) * 100)) & \
            np.array(data['date'] > 20190000 + (i + 1) * 100)
    PM.append(np.array(data[hi].iloc[:, 3:].mean().T))

PM = pd.DataFrame(PM)
PM.columns = data.columns[3:]

month = list(range(1, 13))
PM.insert(0, "Month", month)
PM.insert(1, "type", "PM2.5")
PM.head()
```

```python
to_csv = pd.concat([PM, mean_value])
```

```python
to_csv.to_csv("monthly_average.csv", index=False)
```
